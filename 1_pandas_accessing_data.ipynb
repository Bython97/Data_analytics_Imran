```python
import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

```python
df.head()
```

```python
df[["job_title_short", "job_location", "job_via"]].iloc[0:10]
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.info()
```

```python
#df = df.loc[:, "salary_rate":"salary_hour_avg"].dropna(subset="salary_rate", inplace=True)
```

```python
median_salary_year = df["salary_year_avg"].median()
median_salary_hour = df["salary_hour_avg"].median()
```

```python
df_filled = df
```

```python
df_filled["salary_year_avg"] = df_filled["salary_year_avg"].fillna(median_salary_year)
df_filled["salary_hour_avg"]  = df_filled["salary_hour_avg"].fillna(median_salary_hour)
```

```python
df_filled.loc[:10, "salary_year_avg":"salary_hour_avg"]
```

```python
df_unique = df_filled
df_unique.drop_duplicates()
```

```python
print("Original DataFrame", len(df_filled))
print("Unique DataFrame", len(df_unique))
print("Rows Removed", len(df_filled) - len(df_unique))
```

```python
df_unique.drop_duplicates(subset=["job_title", "company_name"])
print("Original DataFrame", len(df_filled))
print("Unique DataFrame", len(df_unique))
print("Rows Removed", len(df_filled) - len(df_unique))
```
