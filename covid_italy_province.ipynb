```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
df = pd.read_csv("covid19_italy_province.csv")
df
```

```python
province_highest_covid =df.pivot_table(values = "TotalPositiveCases", index = "ProvinceName", aggfunc = "median").sort_values(by = "TotalPositiveCases", ascending = False).head()
```

```python
province_highest_covid.plot(kind='barh', figsize=(10, 6), title='Top 5 Provinces with Highest Median COVID-19 Cases')
plt.xlabel('Province Name') 
plt.ylabel('Median Total Positive Cases')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
