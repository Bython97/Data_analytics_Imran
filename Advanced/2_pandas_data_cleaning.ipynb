```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
median_salary_year = df["salary_year_avg"].median()
```

```python
meadian_salary_hour = df["salary_hour_avg"].median()
```

```python
df_filled = df
```

```python
df_filled["salary_year_avg"] = df_filled["salary_year_avg"].fillna(median_salary_year)
df_filled["salary_hour_avg"] = df_filled["salary_hour_avg"].fillna(meadian_salary_hour)
```

```python
df_filled.loc[0:10, ["salary_year_avg", "salary_hour_avg"]]
```

REMOVING DUPLICATES

```python
df_unique = df_filled
```

```python
df_unique.drop_duplicates()
```

```python
print("The lenght of df_filled is:", len(df_filled))
print("The lenght of df_unique is:", len(df_unique))
print("The number of duplicates removed is:", len(df_filled) - len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset=["job_title", "company_name"])
print("The lenght of df_filled is:", len(df_filled))
print("The lenght of df_unique is:", len(df_unique))
print("The number of duplicates removed is:", len(df_filled) - len(df_unique))
```
